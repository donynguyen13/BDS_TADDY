```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
from matplotlib import pyplot as plt

# Read Data and cleand data
df_oj = pd.read_csv("../examples/oj.csv")

brands = df_oj['brand'].unique()
df_oj['log_sales'] = np.log(df_oj['sales'])
df_oj['log_price'] = np.log(df_oj['price'])
print(df_oj.head())
df_oj_reg = pd.get_dummies(df_oj, columns=['brand'], prefix='brand', drop_first=True)


# Log-Log Regression
x = df_oj_reg[['brand_minute.maid', 'brand_tropicana', 'log_price']]
x = sm.add_constant(x)
y =  df_oj_reg['log_sales']
reg = sm.GLM(y,x).fit()
preds = reg.predict(x) 
df_oj_reg['predicted_log_sales'] = preds

print(reg.summary())
# Regression results don't show dominicks column because only 3 cols are needed to distinguish what we are referencing. 
# If we didn't want dominicks to be the reference column then we need to relevel (R function) the reference column
# The intercpt value gives the value for Dominick's log sales at a log price of zero


# Plot the regression lines on the original data
colors = ['gold', 'red', 'green']
df_oj_reg_viz = df_oj_reg
df_oj_reg_viz['brand_dominicks'] =  np.where((df_oj_reg['brand_minute.maid']==0) & (df_oj_reg['brand_tropicana']==0), 1, 0)
for i, brand in enumerate(brands):
    df_brand = df_oj_reg_viz[df_oj_reg_viz['brand_' + brand] == 1]
    plt.scatter(df_brand['log_price'], df_brand['log_sales'], s=0.2, color=colors[i], label=brand)
    plt.plot(df_brand['log_price'], df_brand['predicted_log_sales'], color=colors[i], linewidth=2)

# Set labels and title
plt.xlabel('Log Price')
plt.ylabel('Log Sales')
plt.title('Log-Log Regression Model W/O Interacton')

# Show the legend
plt.legend()

# Show the plot
plt.xlim([-0.75, 1.5])
plt.show()


# Regression with interaction
df_oj_reg_int = pd.get_dummies(df_oj, columns=['brand'], prefix='brand', drop_first=True)
x = df_oj_reg_int[['brand_minute.maid', 'brand_tropicana', 'log_price']]
x['brand_minute.maid:log_price'] = df_oj_reg_int['brand_minute.maid']*df_oj_reg_int['log_price']
x['brand_tropicana:log_price'] = df_oj_reg_int['brand_tropicana']*df_oj_reg_int['log_price']
x = sm.add_constant(x)
y = df_oj_reg_int['log_sales']
reg_int = sm.GLM(y,x).fit()
preds_int = reg_int.predict(x) 
df_oj_reg_int['predicted_log_sales'] = preds_int

print(reg_int.summary())

# Plot the regression lines on the original data
colors = ['gold', 'red', 'green']
df_oj_reg_int_viz = df_oj_reg_int
df_oj_reg_int_viz['brand_dominicks'] =  np.where((df_oj_reg_int['brand_minute.maid']==0) & (df_oj_reg_int['brand_tropicana']==0), 1, 0)
df_oj_reg_int_viz['brand_dominicks:log_price'] =  np.where((df_oj_reg_int_viz['brand_minute.maid']==0) & (df_oj_reg_int_viz['brand_tropicana']==0), df_oj_reg_int_viz['brand_dominicks']*df_oj_reg_int_viz['log_price'], 0)
for i, brand in enumerate(brands):
    df_brand = df_oj_reg_int_viz[df_oj_reg_int_viz['brand_' + brand] == 1]
    plt.scatter(df_brand['log_price'], df_brand['log_sales'], s=0.2, color=colors[i], label=brand)
    plt.plot(df_brand['log_price'], df_brand['predicted_log_sales'], color=colors[i], linewidth=2)

# Set labels and title
plt.xlabel('Log Price')
plt.ylabel('Log Sales')
plt.title('Log-Log Regression Model With Interacton')

# Show the legend
plt.legend()

# Show the plot
plt.xlim([-0.75, 1.5])
plt.show()
print(df_oj_reg_int.head())


# Regression with interaction and new variable feat
df_oj_reg_int = pd.get_dummies(df_oj, columns=['brand'], prefix='brand', drop_first=True)
x = df_oj_reg_int[['brand_minute.maid', 'brand_tropicana', 'log_price', 'feat']]
x['brand_minute.maid:log_price'] = df_oj_reg_int['brand_minute.maid']*df_oj_reg_int['log_price']
x['brand_tropicana:log_price'] = df_oj_reg_int['brand_tropicana']*df_oj_reg_int['log_price']
x['brand_minute.maid:feat'] = df_oj_reg_int['brand_minute.maid']*df_oj_reg_int['feat']
x['brand_tropicana:feat'] = df_oj_reg_int['brand_tropicana']*df_oj_reg_int['feat']
x['brand_minute.maid:log_price:feat'] = df_oj_reg_int['brand_minute.maid']*df_oj_reg_int['feat']*df_oj_reg_int['log_price']
x['brand_tropicana:log_price:feat'] = df_oj_reg_int['brand_tropicana']*df_oj_reg_int['feat']*df_oj_reg_int['log_price']
x['log_price:feat'] = df_oj_reg_int['feat']*df_oj_reg_int['log_price']
x = sm.add_constant(x)
y = df_oj_reg_int['log_sales']
reg_int = sm.GLM(y,x).fit()
preds_int = reg_int.predict(x) 
df_oj_reg_int['predicted_log_sales'] = preds_int

print(reg_int.summary())


```

```python
# Regression With Time and Date
df_al = pd.read_csv("../examples/airline.csv")
# print(df_al.head())

# Clean data
df_al.columns = df_al.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df_al['year'] = '19' + df_al['year'].astype('str')
df_al['year'] = df_al['year'].astype('int')
df_al['date'] = pd.to_datetime(df_al[['year', 'month']].assign(day=1))

print(df_al.head())

# Graph of monthly passengers by date
plt.plot(df_al['date'], df_al['passengers'], marker='o')

# Formatting the x-axis as dates
plt.xticks(rotation=45)
plt.xlabel('Date')

# Adding labels for clarity
plt.ylabel('Monthly Passengers')
plt.title('Monthly Passengers Over Time')

# Display the graph
plt.show()


# Log scale number of passengers to fix heteroskedasticity
df_al['log_passengers'] = np.log(df_al['passengers'])

# Create new graph with log transformation
plt.plot(df_al['date'], df_al['log_passengers'], marker='o')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Log Monthly Passengers')
plt.title('Log Monthly Passengers Over Time')

plt.show()

# Create regression model
df_al['time'] = (df_al['year']-df_al['year'].min())*12 + df_al['month']
df_al_reg = pd.get_dummies(df_al, columns=['month'], prefix='month', drop_first=True)
dummies = []
for col in df_al_reg.columns:
    if col.startswith('month'):
        dummies.append(col)

X = df_al_reg[['time']].copy()
for i in dummies:
    X.loc[:, i] = df_al_reg[i]
X = sm.add_constant(X)
y = df_al_reg['log_passengers']
al_reg = sm.GLM(y,X).fit()
al_preds = al_reg.predict(X)
df_al_reg['predicted_log_passengers'] = al_preds

print(al_reg.summary())

# Graph of predicted values
plt.plot(df_al_reg['date'], df_al_reg['predicted_log_passengers'], color='red', linestyle='--')

# Oringinal Data
plt.plot(df_al['date'], df_al['log_passengers'], color='blue') 

# Format graph
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Log Monthly Passengers')
plt.title('Predicted vs Original Log Monthly Passengers Over Time')


plt.legend()
plt.show()

# Plot residuals
from statsmodels.graphics.tsaplots import plot_acf

df_al_reg['residuals'] = df_al_reg['log_passengers'] - df_al_reg['predicted_log_passengers']
plt.plot(df_al_reg['time'], df_al_reg['residuals'])
plt.xlabel('time')
plt.ylabel('residuals')
plt.show()
### Plot shows that previous value heavily influences next value which is unsurprising because its time. Like yesterday's weather
### is the best predictor for tomorrow's forecast

print('Autocorrelation with lag 1:', df_al_reg['residuals'].autocorr(lag=1))
plot_acf(df_al_reg['residuals'], lags=20)
plt.xlabel('lag')
plt.ylabel('acf')
plt.show()


# Autoregressive Model
## Clean Data
lag = df_al['log_passengers'].iloc[:-1].tolist()
passengers = df_al['passengers'].iloc[1:].tolist()  # need to add .tolist() because it slicing from df will start index at 1 instead of 0
log_passengers = np.log(passengers)
month = df_al['month'].iloc[1:].tolist()
time = df_al['time'].iloc[1:].tolist()
### Combine elements in dataframe
df_al_areg = pd.DataFrame({
    'time': time,
    'month': month,
    'passengers': passengers,
    'lag': lag,
    'log_passengers': log_passengers
})

    
## Autoregressive model
### Store covariates in data frame X
X = df_al_areg[['time', 'month', 'lag']]
### Create dummies for month column
dummies = []
dummies = pd.get_dummies(df_al_areg, columns=['month'], prefix='month', drop_first=True)
for col in dummies.columns:
    if col.startswith('month'):
        dummies.append(col)
for i in dummies:
    X.loc[:, i] = dummies[i]
X = sm.add_constant(X)
y = df_al_areg['log_passengers']
al_areg = sm.GLM(y,X).fit()
al_apreds = al_areg.predict(X)
# df_al_areg['predicted_log_passengers'] = al_apreds

print(al_areg.summary())
```
